```python
import pandas as pd
import numpy as np
```

```python
#Generar DataFrame desde un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Sofia'],
    'Edad': [23, 25, 30, 32],
    'Ciudad': ['Santiago', 'Talca', 'Temuco', 'Santiago']
}

print(type(data))
```

```python
df = pd.DataFrame(data)
print(type(df))
```

```python
#Cambiar datos
df2 = df.copy()
```

```python
df2.loc[2, 'Edad'] = 99
print(df2)
```

```python
df2.iloc[3, 2] = 'Concepcion'
print(df2)
```

```python
df2['Ciudad'] = 'Conce'
print(df2)
```

```python
df2['Edad2'] = df2['Edad']*2
print(df2)
```

```python
#editar y eliminar columnas 
df2.rename(columns = {'Ciudad': 'Estado'}, inplace=True)
print(df2)
```

```python
df2.drop(columns = ['Edad2'], inplace=True)
print(df2)
```
